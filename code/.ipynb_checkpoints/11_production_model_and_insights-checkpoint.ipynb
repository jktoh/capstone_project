## Import Libraries And Modules

```python
# Basic libraries
import numpy as np
import pandas as pd

# Plotting libraries
import seaborn as sns
```

---

## Load Data

Load `to_preprocess.csv`, `predict.csv` and `feature_importance.csv` from `data` folder into a pandas DataFrame.

```python
df = pd.read_csv('../data/to_preprocess.csv')
predict = pd.read_csv('../data/predict.csv')
feature_importance = pd.read_csv('../data/feature_importance.csv')
```

---

## Plotting And Interpreting Best Model

```python
feature_importance.iloc[:50].plot.barh(
    x='Features', 
    figsize=(20, 15)
);
```

```python
df['log_Common Stock Value'] = np.log(
    df['Common Stock Value']
)
df['log_Goodwill'] = np.log(
    df['Goodwill']
)
df['log_Share Based Compensation'] = np.log(
    df['Share Based Compensation']
)
df['log_Weighted Avg No Of Shares O/S'] = np.log(
    df['Weighted Average Number Of Shares Outstanding Basic']
)
```

```python
ax = sns.pairplot(df, hue='cap', vars=[
    'log_Common Stock Value',
    'log_Goodwill',
    'log_Share Based Compensation',
    'log_Weighted Avg No Of Shares O/S'
])
ax.fig.set_size_inches(15, 15);
```
